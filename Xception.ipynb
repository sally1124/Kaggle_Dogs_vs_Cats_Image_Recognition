```python
img_size = 224

# Build a function to read and process the images to feet the model
# The function returns the data list
def process_data(path,flag):
    data=[]
    for img in os.listdir(path):
            img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
            new_array = cv2.resize(img_array, (img_size, img_size))
            if (flag):
                data.append([new_array,img[0:3]])
            else:
                data.append(new_array)
    return data
```

```python
# load the train data
train_path='/content/Dataset/train'
train=process_data(train_path,True)
train[0]
```

```python
# load the test data
test_path='/content/Dataset/test'
test=process_data(test_path,True)
test[0]
```

# Data Preprocessing

```python
X = np.array([i[0] for i in train]).reshape(-1,img_size,img_size,3)
y = np.array([1 if i[1]=='dog' else 0 for i in train])
```

```python
# Split train data into train and validation
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# Use ImageDataGenerator to preprocess data
from tensorflow.keras.preprocessing.image import ImageDataGenerator

batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255,
                                    rotation_range=20,
                                    width_shift_range=0.2,
                                    height_shift_range=0.2,
                                    shear_range=0.2,
                                    zoom_range=0.2,
                                    horizontal_flip=True,
                                   fill_mode='nearest')

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow(X_train,
                                     y_train,
                                     batch_size=batch_size,
                                     seed = 42)

val_generator = train_datagen.flow(X_val,
                                   y_val,
                                   batch_size=batch_size,
                                   seed = 42)
```

# Build CNN model with transfer learning using Xception

## Prepare the model

```python
# not include the ImageNet classifier at the top
base_model = keras.applications.Xception(
    weights="imagenet",  # Load weights pre-trained on ImageNet.
    input_shape=(img_size, img_size, 3),
    include_top=False,
)
```

```python
# Freeze the base_model
base_model.trainable = False

# Create new model on top
inputs = keras.Input(shape=(img_size, img_size, 3))

# Pre-trained Xception weights requires that input be scaled from (0, 255) to a range of (-1., +1.)
scale_layer = keras.layers.Rescaling(scale=1 / 127.5, offset=-1)
x = scale_layer(inputs)

# The base model contains batchnorm layers.
x = base_model(x, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = keras.layers.Dense(1)(x)
model = keras.Model(inputs, outputs)

model.summary(show_trainable=True)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Train the model

```python
xception_model = model.fit(train_generator,
                  batch_size=batch_size,
                  epochs=12,
                  verbose=1,
                  validation_data=val_generator)
```
